```
import matplotlib.pyplot as plt
import numpy as np
import random as random

class CoinBetGame:
  def __init__(self, p_heads, theta=0.0001):
        self.p_heads = p_heads
        self.theta = theta
        self.rewards = np.zeros(101)
        self.rewards[100] = 1
        self.V = np.zeros(101)

  def lookahead(self, s):
        A = np.zeros(101)
        stakes = range(1, min(s, 100-s)+1)
        for a in stakes:
            A[a] = self.p_heads * (self.rewards[s+a] + self.V[s+a])\
             + (1-self.p_heads) * (self.rewards[s-a] + self.V[s-a])
        return A

  def execute(self):
     while True:
        delta = 0
        for s in range(1, 100):
            A = self.lookahead(s)
            best_action_value = np.max(A)
            delta = max(delta, np.abs(best_action_value - self.V[s]))
            self.V[s] = best_action_value        

        if delta < self.theta:
            break
     policy = np.zeros(100)
     for s in range(1, 100):
         A = self.lookahead(s)
         best_action = np.argmax(A)
         policy[s] = best_action
    
     return policy, self.V

class CoinBetGameModified(CoinBetGame):
  def __init__(self, p_heads, theta=0.0001):
        super(CoinBetGameModified, self ).__init__(p_heads, theta)
        self.rewards = np.ones(101)*-0.0001
        self.rewards[100] = 1
```

```
bet_game = CoinBetGame(0.25)
policy, v = bet_game.execute()

print("")
print("P head: 0.25")
print("")

x = range(100)
y = policy
plt.bar(x, y, align='center', alpha=0.5)
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

bet_game_modified = CoinBetGameModified(0.25)
policy1, v1 = bet_game_modified.execute()

y = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.title('Modified')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()


y = policy
y1 = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

y = v[:100]
y1 = v1[:100]
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('V')
plt.show()



bet_game = CoinBetGame(0.5)
policy, v = bet_game.execute()

print("")
print("P head: 0.5")
print("")

x = range(100)
y = policy
plt.bar(x, y, align='center', alpha=0.5)
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

bet_game_modified = CoinBetGameModified(0.5)
policy1, v1 = bet_game_modified.execute()


x = range(100)
y = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.title('Modified')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

x = range(100)
y = policy
y1 = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

y = v[:100]
y1 = v1[:100]
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('V')
plt.show()

bet_game = CoinBetGame(0.55)
policy, v = bet_game.execute()

print("")
print("P head: 0.55")
print("")

x = range(100)
y = policy
plt.bar(x, y, align='center', alpha=0.5)
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()


bet_game_modified = CoinBetGameModified(0.55)
policy1, v1 = bet_game_modified.execute()



x = range(100)
y = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.title('Modified')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

x = range(100)
y = policy
y1 = policy1
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('Policy')
plt.show()

y = v[:100]
y1 = v1[:100]
plt.bar(x, y, align='center', alpha=0.5)
plt.bar(x, y1, align='center', alpha=0.5)
plt.title('Modified vs Standart')
plt.xlabel('Capital')
plt.ylabel('V')
plt.show()
```
